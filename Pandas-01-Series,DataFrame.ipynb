### Series 연산
###### Series와 숫자연산

Series객체 연산자(+, -, *, /) 숫자

시리즈 개별 원소에 각각 숫자를 연산하고 결과를 시리즈 객체로 반환한다.

```python
score1 = pd.Series({'국어':30, '영어':45, '수학':12})
score1
```

```python
score1 + 50
```

##### Series와 Series 연산
* Series객체 연산자(+,-,*,/) Seires객체
  - index를 찾아 정렬한 후 같은 index의 data(value)들을 연산

```python
score1 = pd.Series({'국어':100, '영어':80, '수학':90})
score2 = pd.Series({'수학':80, '국어':90, '영어':80})
print(score1)
print(score2)
```

```python
score1 + score2
```

##### Series와 Series 연산(연산 메소드)
* Series객체.add(Series객체, fill_value=0)
* fill_value=0 : NaN을 0으로 대체함

```python
import numpy as np

score1 = pd.Series({'국어':np.nan, '영어':80, '수학':90})
score2 = pd.Series({'수학':80, '국어':90})
print(score1)
print(score2)
```

```python
score1.add(score2)
```

```python
score1.add(score2, fill_value=0)
```

```python
score1.add(score2, fill_value=50)
```

### DataFrame - index, columns  변경하기
- **DataFrame.index = [새인덱스리스트]** : 전체 행인덱스 변경
- **DataFrame.columns = [새컬럼명리스트]** : 전체 컬럼명 변경

- **DataFrame.rename(index={}, columns={}, inplace)** : 선택하여 변경 가능
    - key : 기존값
    - value : 바꿀값

  
* **DataFrame.set_index(['열이름'] 또는 '열이름')** : 특정 열을 행인덱스로 설정
* **DataFrame.reset_indet()** : 행인덱스 초기화
    * 행 인덱스를 정수형 위치 인덱스로 초기화한다. 이때 기존 행 인덱스는 열로 이동한다.

```python
import pandas as pd
```

```python
df = pd.DataFrame([['커피', 5000, 60],
                   ['주스', 6000, 45],
                  ['밀크티', 6500, 30]],
                 index = ['coffee', 'juice', 'milktea'],
                 columns = ['제품명', '가격', '판매량'])
df
```

```python
# df의 인덱스를 '음료1', '음료2', '음료3'으로 바꾸기

df.rename(index = {'coffee':'음료1', 'juice':'음료2', 'milktea':'음료3'})
```

```python
df # 원본은 바뀌지 x
```

```python
df.index = ['음료1', '음료2', '음료3']
df
```

```python
# df의 컬럼명을 '음료명', '판매가격', '판매수량'으로 바꾸기

df.rename(columns={'제품명':'음료명', '가격':'판매가격', '판매량':'판매수량'})
```

```python
df
```

```python
df.columns = ['음료명', '판매가격', '판매수량']
df
```

```python
# df의 '음료명'컬럼명을 '제품명'으로 바꾸고, '음료1'인덱스를 'coffee'로 바꾸기

df.rename(columns={'음료명':'제품명'}, index={'음료1':'coffee'}, inplace=True)
df
```

```python
# '제품명' 컬럼을 인덱스로 설정한다.

df.set_index('제품명')
```

```python
df
```

```python
df.set_index(['제품명'], inplace=True)
```

```python
df
```

```python
# df의 인덱스를 정수형 위치 인덱스로 초기화한다.

df.reset_index()
```

# 데이터 입출력

* 판다스는 다양한 형태의 외부 파일을 읽어와서 데이터프레임으로 변환하는 함수를 제공한다.
* 어떤 파일이든 판다스 객체인 데이터프레임으로 변환되고 나면 판다스의 모든 함수와 기능을 자유롭게 사용할 수 있다.
* 반대로 데이터프레임을 다양한 유형의 파일로 저장할 수도 있다.

* csv파일로 저장하기
    * **DataFrame.**<b>`to_csv('파일이름(경로)')`</b>
* csv파일로 읽어오기
    * **pandas.**<b>`read_csv('파일이름(경로)', index_col=인덱스컬럼번호, header=헤더로 지정할 행번호)`</b>
        - **index_col = 0** : 행인덱스 없음 -> 행번호로 인덱스 생성
        - **header = None** : 헤더 없음 -> 열번호로 컬럼명 생성
